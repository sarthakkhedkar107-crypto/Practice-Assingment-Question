```python
def sum_numbers(a, b):
    total = a + b
    print("Sum is:", total)   # Prints the sum
    return total              # Returns the sum

# Function call
result = sum_numbers(10, 20)

print("Returned value:", result)
```

```python
# Global variable
counter = 0

# Function to increment the global counter
def increment():
    global counter
    counter = counter + 1
    print("Counter:", counter)

# Function calls
increment()
increment()
increment()
```

```python
def demo_local():
    message = "Hello, Local"
    print(message)

# Function call
demo_local()
```

```python
def greet(name):
    """
    Function Name: greet

    Purpose:
    Returns a greeting message for the given name.

    Parameter:
    name (str): The name of the person.

    Returns:
    str: A greeting message in the format "Hello, <name>!".
    """
    return "Hello, " + name + "!"

# Function call
print(greet("John"))

# Display the docstring
help(greet)
```

```python
def calculate_area(radius):
    
    pi = 3.14159
    return pi * radius * radius

# Function call
area = calculate_area(5)
print("Area of the circle:", area)

help(calculate_area)
```

```python
def is_even(n):
    if n % 2 == 0:
        return True
    else:
        return False

# Test the function with multiple inputs
print(is_even(10))
print(is_even(7))
print(is_even(24))
print(is_even(15))
```
