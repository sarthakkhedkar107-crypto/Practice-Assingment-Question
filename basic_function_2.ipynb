```python
#Vowel Counter: Create a function count_vowels(word) that loops through a string and prints the number of vowels found.
def count_vowels(word):
    count = 0

    for char in word:
        if char.lower() in "aeiou":
            count += 1

    print("Number of vowels:", count)


count_vowels("Sarthak")
```

```python
#Countdown: Write a function that takes a number x and uses a while loop to count down to zero.
def count(x):
    while x>=0:
        print(x)
        x-=1
        
```

```python
#Simple Calculator: Create a function calc(a, b, operation) where operation can be "add", "sub", "mul", or "div".
def cal(i, j, operation):
    if operation == "add":
        print(i + j)
    elif operation == "sub":
        print(i - j)
    elif operation == "mul":
        print(i * j)
    elif operation == "div":
        print(i / j)
    else:
        print("Invalid operation")

cal(10, 5, "add")
cal(10, 5, "sub")
cal(10, 5, "mul")
cal(10, 5, "div")
```

```python
#Nested Logic: Write a function that takes a list of ages and uses the even_odd logic from your notebook to count how many even-numbered ages exist.
def age(n):
    for i in n:
        if i%2==0:
            print("Even Num",i)
        else:
            print("Odd",i)
```
