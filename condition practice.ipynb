```python
 #  3. Grade Calculator:
 #Write a program to input marks and assign grades as:
 #o 90 and above: "A"
 #o 80-89: "B"
 # o 70-79: "C"
 # o Below 70: "Fail".

marks = float(input('Enter your marks'))

if marks >= 90 and marks <= 100: # True and False --->f
    print('A')
elif marks >= 70 and marks < 90: # True and False
    print('B')
elif marks >= 50 and marks < 70: # T and F
    print('C')
elif marks >= 35 and marks < 50: # T and F
    print('Pass')
elif marks >= 0 and marks < 35: # T and F
    print('Fail')
else:
    print('Please write marks between 0-100 range')
```

```python
# leap year
year = int(input("Enter a year: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

```python
# max Three Number
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if a >= b and a >= c:
    print("Maximum number is:", a)
elif b >= a and b >= c:
    print("Maximum number is:", b)
else:
    print("Maximum number is:", c)
```
